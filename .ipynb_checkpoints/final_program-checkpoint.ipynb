```python
import torch
import glob
import matplotlib.pyplot as plt
from torch.utils.data import Dataset, DataLoader
import torch.nn as nn
import numpy as np
from PIL import Image
from torchvision import transforms, models
from torchvision.datasets import ImageFolder
from torch.optim.lr_scheduler import ReduceLROnPlateau
import torch.optim as optim
from transformers import DistilBertModel, DistilBertTokenizer
import os
import re
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
import seaborn as sns
from torchvision.models import resnet18, ResNet18_Weights

# Enable benchmark mode in cudnn
torch.backends.cudnn.benchmark = True

# Check if GPU is available
device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
print(device)
```

```python
# Paths to datasets
TRAIN_PATH = r"D:\UCalgary\ENEL645\A2\garbage_data\garbage_data\CVPR_2024_dataset_Train"
VAL_PATH = r"D:\UCalgary\ENEL645\A2\garbage_data\garbage_data\CVPR_2024_dataset_Val"
TEST_PATH = r"D:\UCalgary\ENEL645\A2\garbage_data\garbage_data\CVPR_2024_dataset_Test"

# Transforms
torchvision_transform_train = transforms.Compose([
    transforms.Resize((224,224)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomVerticalFlip(),
    transforms.RandomRotation(15),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406] ,std=[0.229, 0.224, 0.225])
])

torchvision_transform_val_test = transforms.Compose([
    transforms.Resize((224,224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406] ,std=[0.229, 0.224, 0.225])
])

# Tokenizer
tokenizer = DistilBertTokenizer.from_pretrained('distilbert-base-uncased')
```

```python
# Function to read image paths, texts, and labels
def get_image_text_label_data(path):
    dataset = ImageFolder(root=path)
    texts = []
    image_paths = []
    labels = []
    for idx in range(len(dataset)):
        image_path, label = dataset.samples[idx]
        # Extract text from filename
        filename = os.path.basename(image_path)
        file_name_no_ext, _ = os.path.splitext(filename)
        text = file_name_no_ext.replace('_', ' ')
        text_without_digits = re.sub(r'\d+', '', text)
        texts.append(text_without_digits)
        image_paths.append(image_path)
        labels.append(label)
    return texts, image_paths, labels

# Prepare datasets
text_train, image_paths_train, labels_train = get_image_text_label_data(TRAIN_PATH)
text_val, image_paths_val, labels_val = get_image_text_label_data(VAL_PATH)
text_test, image_paths_test, labels_test = get_image_text_label_data(TEST_PATH)

class_names = sorted(os.listdir(TRAIN_PATH))
print(class_names)
print("Train set:", len(labels_train))
print("Val set:", len(labels_val))
print("Test set:", len(labels_test))
```

```python
# Define the combined dataset
class CombinedDataset(Dataset):
    def __init__(self, texts, image_paths, labels, image_transform, tokenizer, max_len):
        self.texts = texts
        self.image_paths = image_paths
        self.labels = labels
        self.image_transform = image_transform
        self.tokenizer = tokenizer
        self.max_len = max_len

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        # Load and preprocess image
        image = Image.open(self.image_paths[idx]).convert('RGB')
        image = self.image_transform(image)

        # Tokenize text
        text = str(self.texts[idx])
        label = self.labels[idx]

        encoding = self.tokenizer.encode_plus(
            text,
            add_special_tokens=True,
            max_length=self.max_len,
            return_token_type_ids=False,
            padding='max_length',
            truncation=True,
            return_attention_mask=True,
            return_tensors='pt'
        )

        return {
            'image': image,
            'text': text,
            'input_ids': encoding['input_ids'].flatten(),
            'attention_mask': encoding['attention_mask'].flatten(),
            'label': torch.tensor(label, dtype=torch.long)
        }
```

```python
# Set max_len
max_len = 24

# Create datasets
train_dataset = CombinedDataset(text_train, image_paths_train, labels_train, torchvision_transform_train, tokenizer, max_len)
val_dataset = CombinedDataset(text_val, image_paths_val, labels_val, torchvision_transform_val_test, tokenizer, max_len)
test_dataset = CombinedDataset(text_test, image_paths_test, labels_test, torchvision_transform_val_test, tokenizer, max_len)

# Define batch size and number of workers
batch_size = 64  # set batch size
num_workers = 0

# Create data loaders
trainloader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, num_workers=num_workers, pin_memory=True)
valloader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False, num_workers=num_workers, pin_memory=True)
testloader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False, num_workers=num_workers, pin_memory=True)
```

```python
class GarbageModel(nn.Module):
    def __init__(self, num_classes, input_shape, transfer=False):
        super().__init__()

        self.transfer = transfer
        self.num_classes = num_classes
        self.input_shape = input_shape

        if self.transfer:
            # Load pre-trained ResNet-18 model
            self.feature_extractor = models.resnet18(weights=ResNet18_Weights.IMAGENET1K_V1)
        else:
            # Initialize ResNet-18 with random weights
            self.feature_extractor = models.resnet18(weights=None)

        if self.transfer:
            # Freeze all layers
            for param in self.feature_extractor.parameters():
                param.requires_grad = False
            # Unfreeze last two blocks and fc layer
            for param in self.feature_extractor.layer3.parameters():
                param.requires_grad = True
            for param in self.feature_extractor.layer4.parameters():
                param.requires_grad = True
            for param in self.feature_extractor.fc.parameters():
                param.requires_grad = True

        n_features = self.feature_extractor.fc.in_features
        self.feature_extractor.fc = nn.Linear(n_features, num_classes)
        for param in self.feature_extractor.fc.parameters():
            param.requires_grad = True

    def forward(self, x):
        x = self.feature_extractor(x)
        return x

class DistilBERTClassifier(nn.Module):
    def __init__(self, num_classes):
        super(DistilBERTClassifier, self).__init__()
        self.distilbert = DistilBertModel.from_pretrained('distilbert-base-uncased')
        self.drop = nn.Dropout(0.3)
        self.out = nn.Linear(self.distilbert.config.hidden_size, num_classes)

    def forward(self, input_ids, attention_mask):
        outputs = self.distilbert(input_ids=input_ids, attention_mask=attention_mask)
        hidden_state = outputs.last_hidden_state
        # Mean pooling
        pooled_output = torch.mean(hidden_state, dim=1)
        output = self.drop(pooled_output)
        return self.out(output)
```

```python
# Initialize models
num_classes = len(class_names)
net = GarbageModel(num_classes=num_classes, input_shape=(3,224,224), transfer=True)
net.to(device)

text_model = DistilBERTClassifier(num_classes=num_classes)
text_model.to(device)
```

```python
# Define loss function and optimizers
criterion = nn.CrossEntropyLoss()

# Use mixed precision training
from torch.amp import GradScaler, autocast

# Separate scalers for each model
image_scaler = GradScaler(device='cuda')
text_scaler = GradScaler(device='cuda')

optimizer = torch.optim.AdamW(filter(lambda p: p.requires_grad, net.parameters()), lr=1e-4, weight_decay=1e-4)
scheduler = ReduceLROnPlateau(optimizer, mode='min', patience=2, factor=0.5)

text_optimizer = optim.AdamW(text_model.parameters(), lr=2e-5)
text_scheduler = ReduceLROnPlateau(text_optimizer, mode='min', patience=2, factor=0.5)
```

```python
# Training parameters
nepochs = 20
PATH_IMAGE = './garbage_net.pth'  # Path to save the best image model
PATH_TEXT = './best_text_model.pth'  # Path to save the best text model
early_stopping_wait = 3  # Number of epochs to wait before stopping

# Initialize counters and flags
epochs_no_improve = 0
text_epochs_no_improve = 0
image_training = True
text_training = True

best_loss = 1e+20
best_text_loss = 1e+20
epoch = 0

while (image_training or text_training) and epoch < nepochs:
    epoch += 1
    print(f'\nEpoch {epoch}/{nepochs}')

    # Training Loop for Image Model
    if image_training:
        net.train()
        train_loss = 0.0
        for i, batch in enumerate(trainloader, 0):
            inputs, labels = batch['image'].to(device, non_blocking=True), batch['label'].to(device, non_blocking=True)
            optimizer.zero_grad()
            with autocast(device_type='cuda'):
                outputs = net(inputs)
                loss = criterion(outputs, labels)
            image_scaler.scale(loss).backward()

            # Unscale gradients before optimizer step
            image_scaler.unscale_(optimizer)

            image_scaler.step(optimizer)
            image_scaler.update()
            train_loss += loss.item()
        avg_train_loss = train_loss / len(trainloader)
        print(f'Image Model - Train Loss: {avg_train_loss:.4f}', end=', ')

        # Validation Loop for Image Model
        net.eval()
        val_loss = 0.0
        with torch.no_grad():
            for i, batch in enumerate(valloader, 0):
                inputs, labels = batch['image'].to(device, non_blocking=True), batch['label'].to(device, non_blocking=True)
                outputs = net(inputs)
                loss = criterion(outputs, labels)
                val_loss += loss.item()
        avg_val_loss = val_loss / len(valloader)
        print(f'Val Loss: {avg_val_loss:.4f}')

        # Check for improvement
        if avg_val_loss < best_loss:
            print("Saving best image model")
            epochs_no_improve = 0
            torch.save(net.state_dict(), PATH_IMAGE)
            best_loss = avg_val_loss
        else:
            epochs_no_improve += 1

        if epochs_no_improve == early_stopping_wait:
            print('Early Stopping for Image Model')
            image_training = False

        scheduler.step(avg_val_loss)
    else:
        print('Image Model has stopped training due to early stopping.')

    # Training Loop for Text Model
    if text_training:
        text_model.train()
        total_loss = 0.0
        for batch in trainloader:
            input_ids = batch['input_ids'].to(device, non_blocking=True)
            attention_mask = batch['attention_mask'].to(device, non_blocking=True)
            labels = batch['label'].to(device, non_blocking=True)

            text_optimizer.zero_grad()
            with autocast(device_type='cuda'):
                outputs = text_model(input_ids, attention_mask)
                loss = criterion(outputs, labels)
            text_scaler.scale(loss).backward()

            # Unscale gradients before optimizer step
            text_scaler.unscale_(text_optimizer)

            text_scaler.step(text_optimizer)
            text_scaler.update()
            total_loss += loss.item()
        avg_text_train_loss = total_loss / len(trainloader)
        print(f'Text Model - Train Loss: {avg_text_train_loss:.4f}', end=', ')

        # Validation Loop for Text Model
        text_model.eval()
        val_loss = 0.0
        with torch.no_grad():
            for batch in valloader:
                input_ids = batch['input_ids'].to(device, non_blocking=True)
                attention_mask = batch['attention_mask'].to(device, non_blocking=True)
                labels = batch['label'].to(device, non_blocking=True)
                outputs = text_model(input_ids, attention_mask)
                loss = criterion(outputs, labels)
                val_loss += loss.item()
        avg_text_val_loss = val_loss / len(valloader)
        print(f'Val Loss: {avg_text_val_loss:.4f}')

        # Check for improvement
        if avg_text_val_loss < best_text_loss:
            print("Saving best text model")
            text_epochs_no_improve = 0
            torch.save(text_model.state_dict(), PATH_TEXT)
            best_text_loss = avg_text_val_loss
        else:
            text_epochs_no_improve += 1

        if text_epochs_no_improve == early_stopping_wait:
            print('Early Stopping for Text Model')
            text_training = False

        text_scheduler.step(avg_text_val_loss)
    else:
        print('Text Model has stopped training due to early stopping.')

print('Finished Training')
```

```python
# Load the best models
net = GarbageModel(num_classes=num_classes, input_shape=(3,224,224), transfer=True)
net.load_state_dict(torch.load(PATH_IMAGE, weights_only=True))
net.to(device)
net.eval()

text_model = DistilBERTClassifier(num_classes=num_classes)
text_model.load_state_dict(torch.load(PATH_TEXT, weights_only=True))
text_model.to(device)
text_model.eval()
```

```python
# Combined prediction function
def combined_predict(image_model, text_model, dataloader, device, alpha=0.3):
    image_model.eval()
    text_model.eval()
    all_preds = []
    all_labels = []
    all_image_paths = dataloader.dataset.image_paths
    with torch.no_grad():
        for batch in dataloader:
            images = batch['image'].to(device, non_blocking=True)
            input_ids = batch['input_ids'].to(device, non_blocking=True)
            attention_mask = batch['attention_mask'].to(device, non_blocking=True)
            labels = batch['label'].to(device, non_blocking=True)

            # Image model outputs
            image_outputs = image_model(images)
            image_probs = nn.functional.softmax(image_outputs, dim=1)

            # Text model outputs
            text_outputs = text_model(input_ids, attention_mask)
            text_probs = nn.functional.softmax(text_outputs, dim=1)

            # Combine probabilities
            combined_probs = alpha * image_probs + (1 - alpha) * text_probs

            # Get final predictions
            _, preds = torch.max(combined_probs, dim=1)

            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())
    return np.array(all_preds), np.array(all_labels)
```

```python
# Evaluate on test set
test_preds, test_labels = combined_predict(net, text_model, testloader, device, alpha=0.3)
accuracy = accuracy_score(test_labels, test_preds)
print(f'\nCombined Model Accuracy on Test Set: {accuracy * 100:.2f}%')

# Classification report
report = classification_report(test_labels, test_preds, target_names=class_names)
print(report)
```

```python
# Confusion matrix
cm = confusion_matrix(test_labels, test_preds)

# Plot confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, cmap='Blues', fmt='g', cbar=False, xticklabels=class_names, yticklabels=class_names)
plt.title('Confusion Matrix')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()
```
